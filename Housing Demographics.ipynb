```python
#Dependencies 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Study Data Files 
race_housing_path = "Resources/RaceHousingCity.csv"
age_housing_path = "Resources/AgeHousingCity.csv"
education_housing_path = "Resources/EducationHousingCity.csv"

#Convert to PD
race_housing = pd.read_csv(race_housing_path)
age_housing = pd.read_csv(age_housing_path)
education_housing = pd.read_csv(education_housing_path)
```

```python
#Racial Demographics of Household Owner
race_housing
```

```python
race_housing.columns
```

```python
x =["Philadelphia", "Dallas", "Phoenix", "San Antonio", "San Diego"]
Ywhite =[1289.20, 1328.20, 1059.40, 546, 512.2]
Zblack =[254.1, 211.5, 38.4, 41.2, 20.2]
Yai=[0,0,18.6,9.9,1.4]
Zasain=[82.4, 125.1, 51.7,11,52.9]
Yhis=[97.6, 339, 261.3, 287.5, 101.3]

X_axis=np.arange(len(x))
plt.bar(X_axis -.5, Ywhite, 0.25, align='edge',  label = 'White')
plt.bar(X_axis - .2, Zblack, 0.25,align='edge', label = 'Black')
plt.bar(X_axis +.2, Yai, .25,align='edge', label = 'American Indian or Alaska Native')
plt.bar(X_axis + .3, Zasain, 0.25,align='edge', label = 'Asian')
plt.bar(X_axis + 0, Yhis, 0.25,align='edge', label = 'Hispanic or Latino')

plt.xticks(X_axis, x)
plt.xlabel("City")
plt.ylabel("Racial Demographic of Household Owner")
plt.title("Racial Demographics")
plt.legend()
plt.show()
```
